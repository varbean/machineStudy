```python
import random
u=[i for i in range(20)]
x1=[random.randint(0,10) for i in range(20)]
x2=[random.randint(0,10) for i in range(20)]
x3=[random.randint(0,10) for i in range(20)]
x4=[random.randint(0,10) for i in range(20)]
x5=[random.randint(0,10) for i in range(20)]
x6=[random.randint(0,10) for i in range(20)]
y=[0 for i in range(20)]
for j in range(len(u)):
    for i in range(len(x1)):
        y[i]+=(x1[i]-u[j])**2+(x2[i]-u[j])**2+(x3[i]-u[j])**2+(x4[i]-u[j])**2+(x5[i]-u[j])**2+(x6[i]-u[j])**2
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.plot(u,y)
plt.show()
```
